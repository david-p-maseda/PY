## Create data visualization

### Import plotly library

```python
import plotly.express as px

df_gapminder = px.data.gapminder().query('year == 2007')
df_gapminder
```

### Create scatter plot

Represent the relationship between two **numerical** variables.

```python
table_lifeExp = px.scatter(df_gapminder, x="lifeExp",y="pop")
```

### Change the color

**Exercise:** Based on the continent.

```python
px.scatter(df_gapminder, x="lifeExp",y="pop", color="continent")
```

---

### Change the size

**Exercise:** Based on the population.

```python
px.scatter(df_gapminder, x="lifeExp",y="pop", color="continent",size="pop")
```

### Show point name on hover

**Exercise:** Based on the country name.

```python
px.scatter(df_gapminder, x="lifeExp",y="pop", color="continent",size="pop",hover_name="country")
```

### Create multiple facets

**Exercise:** Based on the continent.

```python
px.scatter(df_gapminder, x="lifeExp",y="pop", color="continent",size="pop",hover_name="country",facet_col="continent") 
```

```python
px.scatter(df_gapminder, x="lifeExp",y="pop", color="continent",size="pop",hover_name="country",facet_col="continent", facet_col_spacing=0.01) 
```
